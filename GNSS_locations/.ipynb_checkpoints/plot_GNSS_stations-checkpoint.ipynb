```python
import pygmt
from os import environ

'''
This code plots the GNSS locations.

To use this code, change the environment to pygmt by running 
"conda activate pygmt" on a new terminal and restart Jupyter notebook.

'''

# concatenate infile name
data_dir = '/Users/oluwaseunfadugba/Documents/Projects/TsE_ValerieDiego/' + \
    'TsE_1D_vs_3D/GNSS_locations'
output_dir = data_dir

infile = data_dir + '/ALL_event.GNSS_locs.txt'
eqs_infile = '/Users/oluwaseunfadugba/Documents/Projects/TsE_ValerieDiego/TsE_1D_vs_3D/Files/events.csv'

# export infile variable as an environment variable
environ['infile'] = infile
environ['eqs_infile'] = eqs_infile

# extract parameters from infile with awk
st_lat_all = !awk 'NR>2 {print $2}' "$infile"
st_lon_all = !awk 'NR>2 {print $3}' "$infile"

eq_lat_all = !awk 'NR>2 {print $2}' "$eqs_infile"
eq_lon_all = !awk 'NR>2 {print $3}' "$eqs_infile"


# ----------  Begin pygmt plot  ------------------

fig = pygmt.Figure()

# Set the region for the plot to be slightly larger than the data bounds.
region = [
    float(min(st_lon_all)) - 1,
    float(max(st_lon_all)) + 1,
    float(min(st_lat_all)) - 1,
    float(max(st_lat_all)) + 1
]
          
fig.coast(
    # Set the x-range from 10E to 20E and the y-range to 35N to 45N
    region=region, #"10/20/35/45",
    # Set projection to Mercator, and the figure size to 15 centimeters
    projection="M15c",
    # Set the color of the land to light gray
    land="gray",
    # Set the color of the water to white
    water="lightblue",
    # Display the national borders and set the pen thickness to 0.5p
    borders="1/0.5p",
    # Display the shorelines and set the pen thickness to 0.5p
    shorelines="1/0.5p",
    # Set the frame to display annotations and gridlines
    frame="ag",
)

fig.basemap(frame=["a", '+t"Map of Japan"'])

fig.plot(x=st_lon_all, 
         y=st_lat_all, 
         style="c0.07c", 
         color="red", 
         pen="black",
         label="GNSS_Stations")

fig.legend()
fig.show()
fig.savefig(data_dir+"fig.GNSS_locations.png")
```
